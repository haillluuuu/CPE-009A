Distance Formula
1. Make a program that would calculate the distance between two points given a list of coordinates. Use the distance  formula.
coorindates_list = [(1,1), (2,3)]
Simple Word Filter


```python
import math

coordinates_list = [(1, 1), (2, 3)]

x1, y1 = coordinates_list[0]
x2, y2 = coordinates_list[1]

distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print("Distance between the points:", distance)
```

3. Create a simple phonebook program that can read from a list of dictionaries. The program should be able to display a  person’s name, contact, and address based from a user input which is the id of the record.


```python
phonebook = [
    {'id': 1, 'name': 'Jenny Lyn Molina', 'contact': '09484062318', 'address': '123 Street, City'},
    {'id': 2, 'name': 'Hail Canoy', 'contact': '09123456789', 'address': '456 Avenue, City'},
    {'id': 3, 'name': 'Toshiro Yagi', 'contact': '08376297936', 'address': 'Musytafu City'}
]

search_id = int(input("Enter the ID to search: "))

found = False
for person in phonebook:
    if person['id'] == search_id:
        print("Name:", person['name'])
        print("Contact:", person['contact'])
        print("Address:", person['address'])
        found = True
        break

if not found:
    print("No record found with ID", search_id)
```
