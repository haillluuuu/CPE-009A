Tasks: A teacher wants to calculate the final grade in a CpE course and want to write it in a python program. The following are the requirements:


1.   PRELIM GRADE = 50% Prelim Exam + 50% Prelim Class Standing (CS)
2.   PRELIM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
3.   MIDTERM GRADE = 1/3 of PRELIM GRADE + 2/3 of (50% Midterm Exam + 50% Midterm Class Standing (CS))
4.   MIDTERM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
5.   FINAL GRADE = 1/3 of MIDTERM GRADE + 2/3 of (50% Final Exam + 50% Final Class Standing (CS))
6.   FINAL CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
7.   HOAs, Quizzes and Assignments are inputted as average of all submissions and are out of 100%.
8.   Major exams are inputted out of 100%.

10.   Provide comments or documentation strings for your program.

```python
print("- - - - - Final Grade Computer - - - - -")

#Prelim Computations
print("\n- - - - - Prelim - - - - - \n")

while stuName > 3:
  stuName = input("Enter Student Name : ")
  hoa = float(input("Enter Prelim Hands-on Activity Grade : "))
  pq = float(input("Enter Prelim Quiz Grade : "))
  pag = float(input("Enter Prelim Assignment Grade : "))
  pcs = hoa*.5 + pq*.3 + pag*.2 #Computes Prelim Class Standing
  print("\nPrelim CS is: ", pcs)
  pe = float(input(("Enter Prelim Exam Grade : ")))
  pg = pe*.5 + pcs*.5 #Computes Prelim Grade

  #Midterm Computations
  print("\n- - - - -  Midterm - - - - - \n")
  mhoa = float(input("Enter Midterm Hands-on Activity Grade : "))
  mpq = float(input("Enter Midterm Quiz Grade : "))
  mag = float(input("Enter Midterm Assignment Grade : "))
  mcs = mhoa*.5 + mpq*.3 + mag*.2 #Computes Midterm Class Standing
  print("\nPrelim CS is: ", mcs)
  me = float(input(("Enter Midterm Exam Grade : ")))
  mg = (1/3 * pg) + (2/3 * (0.5 * me + 0.5 * mcs)) #Computes Midterm Grade

  #Final Computations
  print("\n- - - - - Finals - - - - - \n")
  fhoa = float(input("Enter Final Hands-on Activity Grade : "))
  fpq = float(input("Enter Final Quiz Grade : "))
  fag = float(input("Enter Final Assignment Grade : "))
  fcs = fhoa*.5 + fpq*.3 + fag*.2 #Computes Final Class Standing
  print("\nPrelim CS is: ", fcs)
  fe = float(input(("Enter Final Exam Grade : ")))
  fg = (1/3 * mg) + (2/3 * (0.5 * fe + 0.5 * fcs))

  #Final Outputs
  print("\n- - - - - Final Result - - - - - \n")
  print("\nPrelim Grade is: ", "%.2f" % pg)
  print("\nMidterm Grade is: ", "%.2f" % mg)
  print("\nFinal Grade is: ", "%.2f" % fg)

  if fg >= 75:
    print("Passed")
  if fg >= 90:
    print("Honor Student")
  elif fg >= 95:
    print("With Heighest Honors")
  else:
    print("Failed")
```
