```python
import math

class Shapes:
    def __init__(self, name, sides):
        self.name = name
        self.sides = sides

    def description(self):
        return f"The {self.name} has {self.sides} sides."

    def calculate_area(self, **dimensions):
        name = self.name.lower()
        if name == "triangle":
            return 0.5 * dimensions.get('base', 0) * dimensions.get('height', 0)
        elif name == "circle":
            return math.pi * (dimensions.get('radius', 0) ** 2)
        elif name == "square":
            return dimensions.get('side', 0) ** 2
        elif name == "rectangle":
            return dimensions.get('length', 0) * dimensions.get('width', 0)
        else:
            return 0


obj1 = Shapes("Triangle", 3)
obj2 = Shapes("Circle", 1)
obj3 = Shapes("Square", 4)
obj4 = Shapes("Rectangle", 4)

print(obj1.description())
print(f"Area of {obj1.name}: {obj1.calculate_area(base=10, height=8)}")


print(obj2.description())
area_c = obj2.calculate_area(radius=5)
print(f"Area of {obj2.name}: {area_c:.2f}")


print(obj3.description())
print(f"Area of {obj3.name}: {obj3.calculate_area(side=10)}")


print(obj4.description())
print(f"Area of {obj4.name}: {obj4.calculate_area(length=5, width=10)}")
```

```python
class Shape:

    def __init__(self, name, sides):
        self.name = name
        self.sides = sides

    def description(self):
        return f"The {self.name} has {self.sides} sides."


    def area_circle(self, radius):
        return 3.14 * (radius * radius)

    def area_square(self, side):
        return side * side

    def area_rectangle(self, length, width):
        return length * width

    def area_triangle(self, base, height):
        return 0.5 * base * height


obj1 = Shape("Circle", 0)
obj2 = Shape("Square", 4)
obj3 = Shape("Rectangle", 4)
obj4 = Shape("Triangle", 3)


print(obj1.description())
print(f"Area of {obj1.name}: {obj1.area_circle(5)}")

print(obj2.description())
print(f"Area of {obj2.name}: {obj2.area_square(10)}")

print(obj3.description())
print(f"Area of {obj3.name}: {obj3.area_rectangle(5, 10)}")

print(obj4.description())
print(f"Area of {obj4.name}: {obj4.area_triangle(10, 8)}")



```
