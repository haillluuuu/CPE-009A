For Loop

Example 1

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday" ]
for x in week:
  print (x)
```

Example 2 (The break statement)

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday" ]
for x in week:
  print (x)
  if x == "Thursday":
    break
```

Example 3 (The break statement)

```python
week = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday" ]
for x in week:
  if x == "Thursday":
    break
  print(x)
```

The Range() Function

```python
for x in range(2, 2, 6):
  print(x)
for x in range(2, 6):
  print(x)
```

```python
for x in range(2, 6, 2):
  print(x)
```

Nested Loops

```python
adjective = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adjective:
  for y in fruits:
    print(x, y)
```

Example 5

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

The Continue Statement

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```
