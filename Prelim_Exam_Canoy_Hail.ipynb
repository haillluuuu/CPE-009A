Develop a python scripts that:

Calculate total marks and percentage
Assign grade for each subject based on marks:
90-100: A+

80-89: A

70-79: B+

60-69: B

50-59: C

40-49: D

Below 40: F (Fail)

Determine overall result (Pass/Fail)


Performance Remarks:


Percentage ≥ 90: "Outstanding"

Percentage ≥ 80: "Excellent"

Percentage ≥ 70: "Very Good"

Percentage ≥ 60: "Good"

Percentage ≥ 40: "Satisfactory"

Below 40: "Needs Improvement"

```python
print("- - - - - Grade Calculator - - - - -\n")

grade = int(input("How many subjects? "))

tmarks = 0

for i in range(grade):
    subject_name = input("\nWhat Subject? ")
    mark = int(input(f"whats your mark in {subject_name}? "))

    if mark >= 90:
      print(f"Grade in {subject_name} : A+")
    elif mark >= 80:
      print(f"Grade in {subject_name} : A")
    elif mark >= 70:
      print(f"Grade in {subject_name} : B+")
    elif mark >= 60:
      print(f"Grade in {subject_name} : B")
    elif mark >= 40:
      print(f"Grade in {subject_name} : C")
    else:
      print(f"Grade in {subject_name} : F")

    tmarks += mark
    avg_marks = tmarks / grade

print("\nAverage : ", avg_marks, "\n")

if avg_marks >= 90:
  print("Average Grade : A+ \nPerformance Remarks : Outstanding")
elif avg_marks >= 80:
  print("Average Grade : A \nPerformance Remarks : Excellent")
elif avg_marks >= 70:
  print("Average Grade : B+ \nPerformance Remarks : Very Good")
elif avg_marks >= 60:
  print("Average Grade : B \nPerformance Remarks : Good")
elif avg_marks >= 40:
  print("Average Grade : C \nPerformance Remarks : Satisfactory")
else:
  print("Average Grade : F \nPerformance Remarks : Needs Improvement")
```
