```python

# We want to create a python code for a grading scheme

# Create a function to accept a number of activities
# Compute for the score vs total
# Return the average grade of the student in %
def termcrit(actNumber):
    rawscore = []
    totalscore = []

    for i in range(1, actNumber + 1):
        print(f"\nActivity {i}")

        while True:
            try:
                inputScore = float(input("Score : "))
                inputTotal = float(input("Total : "))

                if inputScore > inputTotal:
                    print("Invalid scores.")
                elif inputScore < 0 or inputTotal <= 0:
                    print("Invalid. Try again.")
                else:
                    break
            except ValueError:
                print("Invalid input. Please enter a number.")

        rawscore.append(inputScore)
        totalscore.append(inputTotal)

    grade = (sum(rawscore) / sum(totalscore)) * 100
    return grade

print("- - - - - Student Grade Calculator - - - - -")
activities = int(input("\nEnter number of activities : "))

average = termcrit(activities)
print(f"\nThe Average Grade of the Student is : {average:.2f}%")
```

```python
# Create a function for the class standing per term


def termCS():
  print("\nHands-on Activities: ")
  termHOA = termCrit(int(input("How many hands-on activities this term? ")))

  print("\nQuizzes:")
  termQuiz = termCrit(int(input("How many quizzes this term? ")))

  print("\nAssignment: ")
  termAss = termCrit(int(input("How many assignments? ")))

  grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
  return grade

print("\n- - - - - Class Standing - - - - -")
final_CS = termCS()
print(f"\nThe Final Class Standing is : {final_CS:.2f}%")
```

```python
# Create a function to compute the term grade
def termGrade():
  print("Computing for the term grade.")
  examRaw = float(input("Exam Score: "))
  examTotal = float(input("Exam Total: "))
  examGrade = (examRaw/examTotal) * 100

  CSgrade = termCS()

  grade = (0.5 * examGrade) + (0.5 * CSgrade)
  return grade
```

```python
# Create a function to compute for the grades in the semester
def semesterGrade():
  print("Prelim Grade")
  prelimGrade = termGrade()

  print("Midterm Grade")
  midtermGrade = termGrade()
  finalMG = (1/3)*prelimGrade + (2/3)*midtermGrade

  print("Final Grade")
  finalGrade = termGrade()
  finalFG = (1/3)*finalMG + (2/3)*finalGrade

  return finalFG
```
